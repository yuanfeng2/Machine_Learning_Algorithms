```python
#(a)
import numpy as np
import time
import matplotlib.pyplot as plt

nvalues = 2000
xvalues = np.linspace(-4,4,nvalues)
yvalues = np.linspace(-4,4,nvalues)
thresh  = 0

# Nonvectorized implementation
t0 = time.time()
f = np.zeros((nvalues,nvalues))

f_thresholded = np.zeros((nvalues,nvalues))
for ix, x in enumerate(xvalues):
    for iy, y in enumerate(yvalues):
        f[ix,iy]             = x**2 - 2 * y**2
        f_thresholded[ix,iy] = f[ix,iy] > thresh
t1 = time.time()
time_nonvectorized = t1 - t0

# Vectorized implementation

X, Y = np.meshgrid(xvalues, yvalues,sparse = True)
t2 = time.time()
f2 = X**2 - 2*Y**2
t3 = time.time()
time_vectorized = t3-t2
# Vectorization speed performance results


print("Time [sec] (non-vectorized): %.4f" % time_nonvectorized)
print("Time [sec] (vectorized):     %.4f" % time_vectorized)
print("The vectorized code is %.4f times faster than the vectorized code" % (time_nonvectorized/time_vectorized))
```

Sermale, indmale and arimale stand for percentage of male workers in service, industry and agriculture respectively. Income is a categorical variable that classifies the countries into four categories(High income, Low income, Upper middle income, and Lower middle income) based on standards posted on the World Bank website.

Using this dataset, we will be able to cross-analyze the percentage of female labor in three different industries worldwide to examine the equality of female workers.

```python
import numpy as np
import pandas as pd
data = pd.read_csv("employment_data.csv")

data.info()
```

(3)

**Data Cleaning**

Missing data exists in this dataset. For columns with the name of "FDI" and "GDPC", a small percentage of the dataset is shown missing. Given the limited amount of missing values, I decided to proceed with the imputations of the missing values using SimpleImputer in sklearn.impute.
Overall, the descriptions of the data suggested no erroneous data exist in the table. The mean, standard deviation, max and min values all appear to be within the reasonable range of values.

The dataset consists of one table and does not require further merging or joining.

```python
data.describe()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer( strategy='mean')
FDI = data.FDI.values.reshape(-1, 1)
imp.fit(FDI)
data_FDI = imp.transform(FDI)
data_FDI
```

```python
GDPC = data.GDPC.values.reshape(-1, 1)
imp.fit(GDPC)
data_GDPC = imp.transform(GDPC)
data_GDPC
```

```python
data["FDI"] = data_FDI
data['GDPC']= data_GDPC
data.info()
```

Another interesting while counter-intuitive finding derived from this EDA is that the previous pattern is not the case when it comes to low-income countries.
